# **Hypothesis Testing**

We will use the t-Test: Paired Two Sample for Means for Hypothesis Testing since the same customers ranked the score for both new product candidates.

Let
<br>
u1: The population mean of Satisfaction score of Orange pen
<br>
u2: The population mean of Satisfaction score of Turquoise pen

Confidence interval 99% (Alpha = 0.01)

H0 : u1 >= u2
<br>
H1 : u1 < u2



image.png

P-value < 0.01, Reject Ho
<br>
Therefore, the population mean of Satisfaction score of Orange pen is less than the Turquiose pen's significantly with 99.9% Confidence interval.
